```python
import random
import numpy as np
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def generateProb(D=1, P=1, Q=1):
    P_mat = np.zeros((P,D))
    Q_mat = np.zeros((Q,D))
    for x in range(P):
        for y in range(D):
            r = random.random()
            if r >= 0.5:
                P_mat[x,y]=1
    for x in range(Q):
        for y in range(D):
            r = random.random()
            if r >= 0.5:
                Q_mat[x,y]=1
    return np.concatenate((Q_mat, P_mat),axis=0)
```

```python
def plotTSNE(D,P,Q):
    X = generateProb(D=D, P=P, Q=Q)
    X_tsne = TSNE(n_components=2,random_state=0).fit_transform(X)
    fig = plt.figure(dpi=200)
    #r = np.random.RandomState(0)
    cmap = plt.cm.get_cmap('gist_ncar', 2 ** D + 1)
    for i in range(2 ** D):
        result = "{:0>{}b}".format(i,D)
        label = result.replace('0','T').replace('1','H')
        result = np.array(list(result), dtype=int)
        points = X_tsne[np.all(X==result, axis=1)]
        plt.scatter(points[:,0], points[:,1], c=[cmap(i/(2**D))], label=label)
    plt.legend(loc='center left', bbox_to_anchor=(1.05, 0.5), fontsize='xx-small')
    plt.title("D={} P={} Q={}".format(D,P,Q))
    plt.show()
```

```python
plotTSNE(2,5,5)
```
